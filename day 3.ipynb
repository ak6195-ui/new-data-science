```python
lis=[]
while True:
    print("1)ADD \n 2)quit")
    OPT=input("ENTER YOUR OPTION \n")
    if OPT.upper().startswith("A"):
        while True:
            ad=(input("ENTER THE NUMBER\n and quit to quit "))
            if ad.upper().startswith("Q"):
                break
            ad=int(ad)
            lis.append(ad)
    elif OPT.upper().startswith("Q"):
        break
    else:
        break
print(list(map(lambda x: f"{x} ->{"even"if x%2==0 else "odd"}",lis)))
```
