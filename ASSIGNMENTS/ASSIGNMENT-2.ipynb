```python
print("WELCOME TO STUDENT MANAGEMENT SYSTEM \n")
def student_result(student):
    for keys,values in student.items():
        print(f"NAME: { keys} \t MARKS:{values["marks"]} \t Grade:{values["grade"]}")
    mark_max=max(student,key=student.get)
    print(f"MAXIMUM:\n NAME:{mark_max} \t MARKS:{student[mark_max]} \n")
    mark_min=min(student,key=student.get)
    print(f"MINIMUM: \n NAME:{mark_min} \t MARKS:{student[mark_min]}\n ")
    markss=[x["marks"] for x in student.values()]
    print(f"AVERAGE: \n {sum(markss)/len(student)}")
student={}
while True:
    print("WHAT OPERATION YOU WANT TO DO??? \n 1)VIEW STUDENT PROFILE \t 2)ADD STUDENT \n")
    op=input()
    if op.upper().startswith("A") or op=="1" :
        print(student_result(student)) 
    elif op.upper().startswith("V") or op=="2":
        while True:
            name=input("ENTER YOUR NAME:\n")
            try:
                marks=int(input("ENTER YOUR MARKS :\n"))
            except Exception as e:
                print("ERROR",e)
            if marks>90 and marks<100:
                grade="A+"
            elif marks>80 and marks<90:
                grade="A"
            elif marks>70 and marks<80:
                grade="B+"
            elif marks>60 and marks<70:
                grade="B"
            else:
                grade="NONE"
                break
            student[name]={"marks":marks,"grade":grade}
            print("DONE \n")
            op2=input("YOU WANT TO ADD MORE??\n OPTIONS:\t 1)YES \t 2)NO")
            if op2.upper().startswith("Y") or op2=="1":
                pass
            else:
                break
    else:
        break
    
            
            
                
```

```python
print("WELCOME TO SHOPPING CART SYSTEM \n")
def shopping_cart(shop):
    print(f"YOUR PRODUCT ARE :\t {shop.keys()}")
    money=[x["price"] for x in shop.values()]
    print(f"YOUR TOTAL AMOUNT IS :\t {sum(money)}")
    dis=[x["discount"] for x in shop.values()]
    print(f"THE TOTAL DISCOUNT IS :{sum(dis)}\n ")
    print(f"TOTAL AMOUNT IS :{sum(money)-sum(dis)}")
shop={}
name=input("ENTER YOUR NAME :\n")
while True:
    print("1)ENTER THE PRODUCT \t 2)VIEW THE BILL \n")
    op=input()
    if op.upper().startswith("E") or op=="1":
        while True:
            products_name=input("ENTER THE PRODUCT NAME :\n")
            try:
                pri=int(input("ENTER THE PRICE :\n"))
            except Exception as e:
                print("ERROR",e)
            if pri>=5000:
                dis=(15/100)*pri
            elif pri>=3000:
                dis=(10/100)*pri
            elif pri>=1000:
                dis=(5/100)*pri
            else:
                dis=0
            shop[products_name]={"price":pri,"discount":dis}
            print("DONE \n")
            print("YOU WANT TO ADD MORE PRODUCTS \n")
            op2=input("1)YES \t 2)NO")
            if op2.upper().startswith("N") or op2=="2":
                break
    elif op.upper().startswith("V") or op=="2":
        print(f"MRS/MS {name} YOUR BILL IS READY :\n")
        shopping_cart(shop)
    else :
        break
```
