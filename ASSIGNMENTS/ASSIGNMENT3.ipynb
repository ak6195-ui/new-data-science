```python
students=["RAM","SHYAM","SITA","HARI","GITA"]
marks=[80,90,91,88,77]
combination=dict(zip(students,marks))
```

```python
def display(combination):
    for i,v in enumerate(combination,start=1):
        print(f"roll no : {i} \t NAME:{v} \t MARKS:{combination[v]}")
def grade(combination):
    for i in combination.keys():
        if combination[i]>90 and combination[i]<100:
            print(f" {i} got GRADE :A+")
        elif combination[i]>80 and combination[i]<90:
            print(f" {i} got GRADE : A")
        elif combination[i]>70 and combination[i]<80:
            print(f"{i} got GRADE : B+")
        else:
            print(f'{i} got FAILL!!!')
def findmax():
    from functools import reduce 
    print(f"{reduce(lambda x,y: x if x>y  else y ,combination.values())} is the highest marks !!!")
        
    
        
```

```python
while True:
    op=input("WHAT DO YOU WANT TO DO :\n 1)DISPLAY \t 2)SHOW GRADE \t 3)FINDMAX \n")
    if op.upper().startswith("D") or op=="1":
        print(display(combination))
    elif op.upper().startswith("S") or op=="2":
        print(grade(combination))
    elif op.upper().startswith("F") or op=="3":
        print(findmax())
    else:
        break
```
