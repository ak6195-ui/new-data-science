```python
print("WELCOME TO STUDENT RECORD MAAGEMENT \n")
di={}
while True :
    print("1)ADD\n 2)VIEW \n")
    op=input()
    if op.upper().startswith("A"):
        name=input("ENTER YOUR NAME : \n")
        age=int(input("ËNTER YOUR AGE \n"))
        grade=int(input("ENTER YOUR GRADE \n"))
        di[name]={age:grade}
        print(di)
    else:
        break
        
```

```python
print("WELCOME TO STUDENT RECORD MANGEMENT \n")
di={}
while True :
    print("1)ADD\n 2)VIEW HIGHEST\n 3) VIEW LOWEST \n 4)VIEW AVERAGE")
    op=input()
    if op.upper().startswith("A") or op=="1":
        name=input("ENTER YOUR NAME : \n")
        marks=int(input("ËNTER YOUR marks \n"))
        di[name]=marks
        print("ADDED SUCESSFULLY \n")
    elif op.upper().startswith("VH") or op=="2":
        maxx=max(di,key=di.get)
        print("STUDENT WITH HIGHEST MARKS IS \n")
        print(maxx,di[maxx])

    elif op.upper().startswith("VL") or op=="3":
        loo=min(di,key=di.get)
        print("THE LOWEST IS \n")
        print(loo,di[loo])
    elif op=="4":
        print("THE AVERAGE MARKS IS:")
        print(sum(di.values())/len(di))
    else:
        break
        
        
        
        
        
        
```
